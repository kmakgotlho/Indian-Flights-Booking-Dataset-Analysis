#What are the airlines in the dataset,accompanied by their frequencies

```python
data.head()
```

```python
#shwoing number of airlines
data['airline'].nunique()
```

```python
# Showing the names of airlines in dataset
data['airline'].unique()
```

```python
# Showing all the airlines with their frequencies
data['airline'].value_counts()
```

```python
# Showing all airlines with their number of frequencies in a bar graph

data['airline'].value_counts(ascending=True).plot.barh(color=['pink','red'])

plt.title('Airlines with frequencies')

plt.xlabel('Number of flights')

plt.ylabel('Airline')

plt.show()
```

```python
#Show departure time
data['departure_time'].value_counts()
```

```python
#Show arrival time for flights
data['arrival_time'].value_counts()
```

```python
#Showing the departure time and arrival time for flights with their counts

plt.figure(figsize =(16,4))

plt.subplot(1,2,1)

plt.bar(data['departure_time'].value_counts().index, data['departure_time'].value_counts().values, color=['pink','red'])
plt.title('Departure Time')
plt.xlabel('D.Time')
plt.ylabel('D.Frequency')

plt.subplot(1,2,2)

plt.bar(data['arrival_time'].value_counts().index, data['arrival_time'].value_counts().values, color=['g','y'])
plt.title('Arrival Time')
plt.xlabel('A.Time')
plt.ylabel('A.Frequency')
plt.show()
```

```python
#source city and destination city

data['source_city'].value_counts()
```

```python
#Showing the destination city of flights
data['destination_city'].value_counts()
```

```python
ZZ#Showing source city and destination city with their counts

plt.figure(figsize=(16,4))

plt.subplot(1,2,1)

plt.barh(data['source_city'].value_counts().index, data['source_city'].value_counts().values, color=['pink','red'])
plt.title("Source Cties with No. of Flights")
plt.xlabel('No. of flights')
plt.ylabel('City')

plt.show()

plt.subplot(1,2,2)

plt.barh(data['destination_city'].value_counts().index, data['destination_city'].value_counts().values, color=['y','g'])
plt.title("Destination Cities with No. of Flights")
plt.xlabel('No. of flights')
plt.ylabel('City')

plt.show()
```
